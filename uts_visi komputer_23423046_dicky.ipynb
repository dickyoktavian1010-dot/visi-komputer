```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

### Menentukan Skala dari Gambar

Untuk menghitung luas lahan, langkah pertama yang krusial adalah menentukan skala gambar. Ini dilakukan dengan mengidentifikasi objek pada denah yang memiliki ukuran sebenarnya yang diketahui. Misalnya, jika ada jalan dengan lebar 5 meter, atau sebuah bangunan dengan panjang tertentu yang Anda ketahui.

```
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Pastikan variabel `image_path` dan `img_pil` tersedia dari sel sebelumnya
# Jika tidak, Anda perlu memuat ulang gambar:
# image_path = '/content/lahan timah.png'
# img_pil = Image.open(image_path)
# img = np.array(img_pil)

# Tampilkan gambar dan minta pengguna untuk memilih dua titik
print("Silakan klik dua titik pada gambar yang merepresentasikan objek dengan ukuran yang Anda ketahui di dunia nyata.")
print("Misalnya, dua ujung dari sebuah pagar, atau lebar jalan.")

fig, ax = plt.subplots(figsize=(10, 8))
ax.imshow(img)
ax.set_title("Klik dua titik untuk kalibrasi skala")
ax.axis('off')

# Menggunakan ginput untuk mendapatkan koordinat dua titik
points = plt.ginput(2, timeout=0) # timeout=0 agar tidak ada timeout otomatis
plt.close(fig)

if len(points) == 2:
    p1 = np.array(points[0])
    p2 = np.array(points[1])

    # Hitung jarak piksel antara dua titik
    pixel_distance = np.linalg.norm(p2 - p1)
    print(f"Jarak piksel antara dua titik yang dipilih: {pixel_distance:.2f} piksel")

    # Minta pengguna untuk memasukkan jarak sebenarnya
    try:
        real_world_distance_str = input("Masukkan jarak sebenarnya dari objek yang Anda pilih (misal: 10 meter, atau 50 cm): ")
        # Pisahkan angka dari unit
        import re
        match = re.match(r"([0-9.]+)\s*([a-zA-Z]+)", real_world_distance_str)
        if match:
            real_world_value = float(match.group(1))
            real_world_unit = match.group(2).lower()

            # Konversi semua ke meter untuk konsistensi, atau unit dasar lainnya
            if real_world_unit == 'cm':
                real_world_value /= 100
                real_world_unit = 'meter'
            elif real_world_unit == 'km':
                real_world_value *= 1000
                real_world_unit = 'meter'
            # Tambahkan konversi unit lain jika diperlukan

            if pixel_distance > 0:
                scale_factor_per_pixel = real_world_value / pixel_distance
                print(f"Skala gambar: 1 piksel = {scale_factor_per_pixel:.4f} {real_world_unit}")
                print(f"Ini berarti, setiap piksel pada gambar merepresentasikan {scale_factor_per_pixel:.4f} {real_world_unit} di dunia nyata.")
            else:
                print("Jarak piksel terlalu kecil untuk menghitung skala.")
        else:
            print("Format input jarak tidak valid. Harap masukkan angka diikuti dengan unit (misal: '10 meter').")

    except ValueError:
        print("Input jarak sebenarnya harus berupa angka.")

else:
    print("Anda tidak memilih dua titik. Skala tidak dapat dihitung.")
```

Setelah skala berhasil dihitung, langkah selanjutnya adalah **segmentasi lahan** dan kemudian **perhitungan luas**. Proses ini akan melibatkan:

1.  **Segmentasi:** Mengidentifikasi dan memisahkan area 'lahan' dari bagian lain gambar (misalnya, jalan, bangunan lain, atau area yang tidak relevan).
    *   Ini bisa dilakukan secara manual (menggunakan alat seperti `plt.ginput` lagi untuk menggambar poligon di sekitar lahan) atau otomatis (menggunakan teknik pemrosesan gambar seperti deteksi warna, deteksi tepi, atau segmentasi berbasis citra).

2.  **Perhitungan Luas Piksel:** Setelah lahan tersegmentasi, kita akan menghitung jumlah piksel yang membentuk area lahan tersebut.

3.  **Konversi ke Luas Dunia Nyata:** Mengalikan jumlah piksel dengan faktor skala kuadrat (karena luas dihitung dalam unit kuadrat, misal `meter^2`) untuk mendapatkan luas sebenarnya.

```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.path import Path
from matplotlib.patches import PathPatch

# Memastikan img tersedia dari sel sebelumnya. Jika tidak, Anda perlu memuat ulang:
# image_path = '/content/lahan timah.png'
# img_pil = Image.open(image_path)
# img = np.array(img_pil)

# --- PENTING: Pastikan scale_factor_per_pixel dan real_world_unit telah diatur dari sel sebelumnya! ---
# Jika belum diatur, silakan jalankan sel penentuan skala (cell 4f934b3d) dan masukkan jarak sebenarnya.
# Untuk contoh ini, saya akan menggunakan nilai placeholder jika belum ada (Anda harus menggantinya):
if 'scale_factor_per_pixel' not in locals():
    print("PERINGATAN: 'scale_factor_per_pixel' tidak ditemukan. Menggunakan nilai placeholder: 1 piksel = 0.1 meter.")
    scale_factor_per_pixel = 0.1 # Placeholder: 1 piksel = 0.1 meter
    real_world_unit = 'meter'
if 'real_world_unit' not in locals():
    real_world_unit = 'meter'

print("\nSekarang, silakan klik titik-titik pada gambar untuk menandai batas luar lahan yang ingin Anda ukur.")
print("Klik kanan atau tombol tengah mouse untuk mengakhiri pemilihan.\n")

fig, ax = plt.subplots(figsize=(12, 10))
ax.imshow(img)
ax.set_title("Klik titik-titik batas lahan (klik kanan/tengah untuk selesai)")
ax.axis('off')

# Menggunakan ginput untuk mendapatkan koordinat titik-titik batas lahan
polygon_points = plt.ginput(n=-1, timeout=0, show_clicks=True) # n=-1 untuk jumlah tak terbatas
plt.close(fig)

if len(polygon_points) > 2: # Minimal 3 titik untuk membuat poligon
    # Konversi list tuple ke numpy array
    vertices = np.array(polygon_points)

    # Membuat Path dari titik-titik
    path = Path(vertices)

    # Untuk menghitung luas piksel di dalam poligon, kita bisa menggunakan mask
    # Membuat grid koordinat untuk seluruh gambar
    ny, nx = img.shape[0], img.shape[1] # Dimensi gambar (tinggi, lebar)
    x, y = np.meshgrid(np.arange(nx), np.arange(ny))
    points_grid = np.vstack((x.flatten(), y.flatten())).T

    # Mengecek piksel mana yang berada di dalam poligon
    mask = path.contains_points(points_grid)
    pixel_area = np.sum(mask)

    print(f"Luas lahan dalam piksel: {pixel_area:.2f} piksel persegi")

    # Konversi luas piksel ke luas dunia nyata
    # Luas = (jumlah piksel) * (skala per piksel)^2
    real_world_area = pixel_area * (scale_factor_per_pixel ** 2)

    print(f"Faktor skala yang digunakan: 1 piksel = {scale_factor_per_pixel:.4f} {real_world_unit}")
    print(f"Luas lahan yang diprediksi: {real_world_area:.2f} {real_world_unit}^2")

    # Opsional: Tampilkan gambar dengan poligon yang digambar
    fig_final, ax_final = plt.subplots(figsize=(12, 10))
    ax_final.imshow(img)
    ax_final.add_patch(PathPatch(path, facecolor='green', alpha=0.5, edgecolor='red', lw=2))
    ax_final.set_title(f"Lahan Tersegmentasi (Luas: {real_world_area:.2f} {real_world_unit}^2)")
    ax_final.axis('off')
    plt.show()

else:
    print("Anda tidak memilih cukup titik untuk membuat poligon (minimal 3 titik diperlukan).")
    print("Silakan jalankan ulang sel ini dan coba lagi.")
```
